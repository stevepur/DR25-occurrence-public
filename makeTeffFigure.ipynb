```python
import numpy as np
import requests
import pandas as pd
from cStringIO import StringIO
import matplotlib.pyplot as plt
```

```python
stellarCatalog = "stellarCatalogs/dr25_stellar_berger2019_clean_GK.txt"
stellarTargets = pd.read_csv(stellarCatalog)
```

```python
plt.hist(stellarTargets.teff, 100);
plt.plot([5777,5777], [0,1400])
```

```python
plt.hist(stellarTargets.radius, 100);
```

```python
plt.plot(stellarTargets.teff, stellarTargets.radius, '.')
```

```python
r = stellarTargets.radius;
t = stellarTargets.teff/5777;
plt.hist(r**2 * t**4, 100, histtype="step", color="k");
plt.tick_params(labelsize = 18)
plt.xlabel("Luminosity ($L_{\odot}$)", fontsize = 24)
plt.grid(axis='x', linestyle='--')
plt.xlim(0,3)
plt.gca().set_yticklabels([])

plt.savefig("stellarLum.pdf",bbox_inches='tight')
```
