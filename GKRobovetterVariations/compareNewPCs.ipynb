```python
combNewPcs = pd.merge(newPCsHiCompleteness, newPCsFpwgpc, on="TCE_ID", how="outer")
print("combNewPcs has " + str(len(combNewPcs)) + " PCs")
```

```python
combNewPcs
```

```python
combNewPcs["koi_period"] = np.where(np.isnan(combNewPcs.koi_period_x), 
                                    combNewPcs.koi_period_y, combNewPcs.koi_period_x)
combNewPcs["corrected_prad"] = np.where(np.isnan(combNewPcs.corrected_prad_x), 
                                    combNewPcs.corrected_prad_y, combNewPcs.corrected_prad_x)
combNewPcs = combNewPcs.sort_values(by=["koi_period"])
```

```python
combNewPcs.to_csv("combNewPcs.csv", 
              columns=["TCE_ID", "kepoi_name", "koi_period", "corrected_prad", "totalReliability", "reliability", "fpp_prob_use"],
             index=False)
```

```python
print("% new PCs in highcompleteness and fpwgpc")
print("\\tablehead{TCE ID & KOI & Period & Radius & HC Reliability & FPWWG PC Reliability}")
for i in range(len(combNewPcs)):
    print(combNewPcs.iloc[i].TCE_ID + " & ", end = '')
    if isinstance(combNewPcs.iloc[i].kepoi_name_x, str):
        print(combNewPcs.iloc[i].kepoi_name_x + " & ", end = '')
    else:
        print(combNewPcs.iloc[i].kepoi_name_y + " & ", end = '')
    print(str(np.round(combNewPcs.iloc[i].koi_period, 2)) + " & ", end = '')
    print(str(np.round(combNewPcs.iloc[i].corrected_prad, 2)) + " & ", end = '')
    if not np.isnan(combNewPcs.iloc[i].koi_period_x):
        print(str(np.round(combNewPcs.iloc[i].totalReliability_x, 2)) + " & ", end = '')
        if not np.isnan(combNewPcs.iloc[i].koi_period_y):
            print(str(np.round(combNewPcs.iloc[i].totalReliability_y, 2)) + " ", end = '')
        else:
            print("- ", end = '')
    else:
        if not np.isnan(combNewPcs.iloc[i].koi_period_x):
            print(str(np.round(combNewPcs.iloc[i].totalReliability_x, 2)) + " & ", end = '')
        else:
            print("- & ", end = '')
        print(str(np.round(combNewPcs.iloc[i].totalReliability_y, 2)) + " ", end = '')
    print("\\\\")
```

```python
cbc = ['#377eb8', '#ff7f00', '#4daf4a',
                  '#f781bf', '#a65628', '#984ea3',
                  '#999999', '#e41a1c', '#dede00']
fig, ax = plt.subplots(figsize=(10,6));
scf = ax.plot(dr25PCs.koi_period, dr25PCs.corrected_prad, 'k.', ms=3, alpha=0.1);
scf = ax.plot(newPCsHiCompleteness.koi_period, newPCsHiCompleteness.corrected_prad, color=cbc[0], marker='o', ms=9, ls="None");
scf = ax.plot(newPCsFpwgpc.koi_period, newPCsFpwgpc.corrected_prad, color=cbc[1], marker='+', mew=2, ms=10, ls="None");
# plt.xlim([50, 400])
plt.legend(["DR25 PCs", "High Completeness PCs", "FPWG PCs"], fontsize = 16, loc=1)
plt.plot([400, 400], [0.75, 2.5], color='k', linestyle=':', linewidth=1)
plt.plot([50, 400], [0.75, 0.75], color='k', linestyle=':', linewidth=1)
plt.plot([50, 400], [2.5, 2.5], color='k', linestyle=':', linewidth=1)
plt.plot([50, 50], [0.75, 2.5], color='k', linestyle=':', linewidth=1)
plt.ylim([0, 5])
plt.tick_params(labelsize = 16)
plt.ylim(0,6)
plt.xlabel("Period", fontsize = 24);
plt.ylabel("Radius", fontsize = 24);
plt.savefig("newPCs.pdf",bbox_inches='tight')
plt.title("new PCs in the low-reliability robovetter result");
```
